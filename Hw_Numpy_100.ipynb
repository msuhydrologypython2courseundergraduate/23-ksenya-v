```python
#44 рассмотрtnm случайную матрицу размером 10х2, представляющую декартовы координаты, преобразуем их в полярные координаты
import numpy as np
a = np.random.randn(10, 2)  # 10 точек, каждая имеет (x, y)

print("Декартовы координаты (x, y):")
print(a)

# Преобразуем в полярные координаты (r, phi)
# r = sqrt(x^2 + y^2)
# phi = arctan2(y, x) - угол в радианах от -pi до pi
r = np.sqrt(a[:, 0]**2 + a[:, 1]**2)
phi = np.arctan2(a[:, 1], a[:, 0])

# Собираем полярные координаты в матрицу
polar_coords = np.column_stack((r, phi)) #np.column_stack — объединяет одномерные массивы в виде столбцов в двумерный массив.
#Преобразуем радианы в градусы для наглядности
phi_degrees = np.degrees(phi)
polar_coords_degrees = np.column_stack((r, phi_degrees))
print(polar_coords_degrees)
```
