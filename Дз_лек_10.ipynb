```python
#Создайте датафрейм 10 на 10, заполненную случайным образом 0 и 1.
#Можно ли выбрать только колонки с четными индексами, не используя iloc или указывая индексы вручную?
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0, 2, size=(10, 10)))
print(df)

# Выбор колонок с четными индексами
even_columns = df[df.columns[df.columns % 2 == 0]]

print(even_columns)
print(even_columns.shape)
```

```python
#Найдите сумму значений в датафрейме
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0, 2, size=(10, 10)))

print(df)

sum = df.values.sum()
print(sum)
```

```python
#Замените 0 и 1 на "ноль" и "один"
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0, 2, size=(10, 10)))

print(df)
# Замена 0 → "ноль", 1 → "один"
df_replaced = df.replace({0: "ноль", 1: "один"})
print(df_replaced)
```
